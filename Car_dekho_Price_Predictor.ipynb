# MODEL

```python
X = df1.drop(columns = 'Selling_Price')
X
```

```python
Y = df1['Selling_Price']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['Car_Name',"company","Fuel_Type","Transmission"]])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['Car_Name',"company","Fuel_Type","Transmission"]),remainder = 'passthrough')
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

```python
scores = []
for i in range(1000):
  X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(X_train,Y_train)
  Y_pred = pipe.predict(X_test)
  scores.append(r2_score(Y_test,Y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state =np.argmax(scores) )
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
Y_pred = pipe.predict(X_test)
r2_score(Y_test,Y_pred)
```
